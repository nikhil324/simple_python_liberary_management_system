```python
from flask import Flask
from flask_sqlalchemy import SQLAlchemy
from flask_login import UserMixin,LoginManager,login_user
from datetime import datetime
app =Flask(__name__)
app.config['SQLALCHEMY_DATABASE_URI']='sqlite:///liberary.db'
app.secret_key="nikhil@123" 
db=SQLAlchemy(app)
login_manager=LoginManager()
login_manager.init_app(app)

class lib(UserMixin,db.Model):
    id = db.Column(db.TEXT(50), primary_key=True)
    books = db.Column(db.TEXT(50))
    Return_book = db.Column(db.TEXT(50))
    date=db.Column(db.TEXT(10),index = True,default=datetime.now) 
def signin(id,book):
    userid=id
    notes=book
    sob = lib(id=userid,books=notes)
    db.session.add(sob)
    db.session.commit()
def ret(id_no,book):
    re= lib(id=id_no,Return_book=book)
    db.session.add(re)
    db.session.commit()
    
class local(UserMixin,db.Model):
    name = db.Column(db.TEXT(50), primary_key=True)
    book = db.Column(db.TEXT(50))
    address=db.Column(db.TEXT(50))
    security_money = db.Column(db.INTEGER)
    date=db.Column(db.TEXT(10),index = True,default=datetime.now) 
    non_refendable=db.Column(db.INTEGER)
    return_money=db.Column(db.INTEGER)
    
def locale(name,book,address,nonrefendable):
    name=name
    address=address
    notes=book
    nonrefendable=nonrefendable
    rob = local(name=name,book=notes,non_refendable=nonrefendable,address=address)
    db.session.add(rob)
    db.session.commit()
def locala(name,book,address,security_money):
    name=name
    address=address
    notes=book
    security_money=security_money
    dob = local(name=name,book=notes,security_money=security_money,address=address)
    db.session.add(dob)
    db.session.commit()    
class accounts(UserMixin,db.Model):
    id_no=db.Column(db.TEXT(50),primary_key=True)
    password=db.Column(db.TEXT(50))
    fees=db.Column(db.INTEGER)
    date=db.Column(db.TEXT(10),index = True,default=datetime.now)     
def accts(new_id,pas,fees):
    acc=accounts(id_no=new_id,password=pas,fees=fees)
    db.session.add(acc)
    db.session.commit()
class books(UserMixin,db.Model):
    books = db.Column(db.TEXT(50), primary_key=True)
def localr(book_rate,name,address,book):
    lo=local(return_money=book_rate,name=name,address=address,book=book)
    db.session.add(lo)
    db.session.commit()
    

    
    
if __name__ == '__main__':          
    app.run(debug=True) 
```
